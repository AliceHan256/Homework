### Index Selection Using Loc

```python
people_csv.head(2)
```

```python
# Set the index as the 'first_name' column
people_csv = people_csv.set_index(["first_name"])
people_csv.head(2)
```

```python
# Set multiple index as the 'first_name' as well as 'last_name' columns
people_csv_index1 = people_csv.set_index(['last_name', 'gender'])
people_csv_index1.head(2)
```

```python
# Sort the index
people_csv = people_csv.sort_index()
people_csv.head()
```

```python
# Select the row with the index 'Evan'
people_csv.loc["Abraham"]
```

```python
# Slice the data to output a range of rows based on the index
people_csv.loc["Aleshia":"Svetlana"]
```

```python
# Filter rows based on a column value conditional
people_csv.loc[people_csv['gender'] == 'M'].head()
```

```python
# Modify the 'first_name' value of the row with the index 'Yun'
people_csv.loc['Yun', 'first_name'] = 'Yuna'
people_csv.head()
```

```python
# Filter rows based on a column value conditional
people_csv.loc[people_csv["gender"] == "M"].head()
```

### Assignment Using Loc

```python
# Modify the 'first_name' value of the row with the index 'Yun'
print("Original Data")
print("*"*20)
people_csv.loc["Yun"]
```

```python
# 'first_name' modification
people_csv.loc["Yun", "first_name"] = "Yuna"
```
